```python
pokemon["Total"] = pokemon["HP"] + pokemon["Attack"] + pokemon["Defense"] + pokemon["Sp. Atk"] + pokemon["Sp. Def"] + pokemon["Speed"]
pokemon.head(5)

# Dropping columns
pokemon.drop(columns=["Total"])
```

```python
# Adding columns a different way
pokemon.head(5)

pokemon["Total"] = pokemon.iloc[:,4:10].sum(axis=1)
pokemon.head(5)

# Moving Total from end to the middle

cols = list(pokemon.columns.values)
pokemon = pokemon[cols[0:4] + [cols[-1]] + cols[4:12]]
pokemon
```

```python
# Saving our dataframe

pokemon.to_csv("totalPokemon.csv", index=False)

# Saving as .xlsx
pokemon.to_excel("totalPokemon.xlsx", index=False)
```

```python
# Filtering data

pokemon.loc[pokemon["Type 1"] == "Grass"]
```

```python
# Filtering multiple columns with AND

pokemon.loc[(pokemon["Type 1"] == "Grass") & (pokemon["Type 2"] == "Fighting")]
```

```python
# Filtering using OR

pokemon.loc[(pokemon["Type 1"] == "Ghost") | (pokemon["Type 2"] == "Fighting")]
```

```python
# Filtering multiple columns with AND as well as filtering using char values and integers

pokemon.loc[(pokemon["Type 1"] == "Water") & (pokemon["Type 2"] == "Psychic") & (pokemon["HP"] >= 70)]
```

```python
# Adding to new dataframe

water_psychic = pokemon.loc[(pokemon["Type 1"] == "Water") & (pokemon["Type 2"] == "Psychic") & (pokemon["HP"] >= 70)]

# Let's reset the index
water_psychic.reset_index(drop=True, inplace=True)
water_psychic
```

```python
# Removing certain character names.

# The "~" is the "Not" function. Meaning it calls for everything that does NOT have what you state in the statement.

noMega = pokemon.loc[~pokemon["Name"].str.contains("Mega")]
noMega

noMega.to_csv("noMega.csv", index=False)
```

```python
# Looking for just Grass or Fire pokemon using regex:

pokemon.loc[pokemon["Type 1"].str.contains("Grass|Fire", regex=True)]

# using flags=re.I, if you type in "grass|fire", it automatically capitalizes the query for you to correctly search
```

```python
# All pokemon names that start with "Pi" using regex (Remember to import re)

pokemon.loc[pokemon["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)]
```

```python
# Conditional changes

# Changing ALL Fire types to "Volcano" type

pokemon.loc[pokemon["Type 1"] == "Fire", "Type 1"] = "Volcano"
pokemon.head(10)
```

```python
# Changing ALL Grass types to Legendary = True

pokemon.loc[pokemon["Type 1"] == "Grass", "Legendary"] = True
pokemon.head(10)
```

```python
# Reloading the dataframe because you fucked with it too much

pokemon = pd.read_csv("totalPokemon.csv")
pokemon.head(10)
```

```python
# Aggregate statistics: GROUPBY

# Are specific types prone to having high Attack?

pokemon.groupby(["Type 1"]).mean().sort_values("Attack", ascending=False)
```

```python
# Some other ways of manipulating data:

# .sum()
# .count()

# Count example:

pokemon["Count"] = 1

typeBreakdown = pokemon.groupby(["Type 1", "Type 2"]).count()["Count"]
typeBreakdown = pd.DataFrame(typeBreakdown)
typeBreakdown

typeBreakdown.to_csv("test.csv", index=True)
```
